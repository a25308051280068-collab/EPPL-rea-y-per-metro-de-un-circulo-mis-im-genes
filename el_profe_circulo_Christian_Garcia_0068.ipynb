# Formula para calcular area
$A=\pi*r^2$
----

## Formula para calcular perimetro
$P=2*\pi*r$
---
## Datos de entrada
$r$ ?
---

```python
# calcula area y perimetro de un circulo
# Variable A, P, r
# Formula A = pi*r^2 y P = 2 * pi * r
# capturar el valor de r
import math
print("Calcular area perimetro de un circulo")
r = float(input("Ingrese el valor de r: "))
A = math.pi *r**2
P = 2 * math.pi * r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Christian Garcia NL = 21")
```
